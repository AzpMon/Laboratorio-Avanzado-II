```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
azucar_concentracion = pd.read_excel("Azucar_concentracion_variable.xlsx")
azucar_concentracion.head()
```

```python
X_azucar_concentracion= sm.add_constant(azucar_concentracion["θ_grados"])
Y_azucar_concentracion= azucar_concentracion["Concentración_gms"]
modelos_azucar_concentracion = sm.OLS(Y_azucar_concentracion,X_azucar_concentracion).fit()
print(modelos_azucar_concentracion.summary())
```

```python
data = np.linspace(88, 102, 100)
plt.scatter(azucar_concentracion.θ_grados, azucar_concentracion.Concentración_gms, color = "black", label = "Datos")
plt.plot(data, -584.6242  +  6.5547*data, c = "black" ,label = "Ajuste", linestyle = "--")
plt.title("Actividad óptica azúcar: $\\lambda = 532nm$, 500ml $H_2 O$, L=8.944")
plt.xlabel("$\\theta[grados]$"); plt.ylabel("Concentración [gms]"); plt.grid(); plt.legend()
```

```python
miel_karo_concentracion = pd.read_excel("Miel_karo_concentracion_variable.xlsx")
X_miel_karo_concentracion= sm.add_constant(miel_karo_concentracion["θ_grados"])
Y_miel_karo_concentracion= miel_karo_concentracion["Concentración_gms"]
modelos_miel_karo_concentracion = sm.OLS(Y_miel_karo_concentracion,X_miel_karo_concentracion).fit()
print(modelos_miel_karo_concentracion.summary())
```

```python
data = np.linspace(88, 108, 100)
plt.scatter(miel_karo_concentracion.θ_grados, miel_karo_concentracion.Concentración_gms, color = "blue", label = "Datos")
plt.plot(data, -347.6454   +  3.8946*data, c = "blue" ,label = "Ajuste", linestyle = "--")
plt.title("Actividad óptica Miel Karo: $\\lambda = 532nm$, 500ml $H_2 O$, L = 8.954cm")
plt.xlabel("$\\theta[grados]$"); plt.ylabel("Concentración [gms]"); plt.grid(); plt.legend()
```
