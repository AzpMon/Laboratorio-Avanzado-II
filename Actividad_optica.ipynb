```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
azucar_concentracion = pd.read_excel("Azucar_concentracion_variable.xlsx")
azucar_concentracion.head()
```

```python
X_azucar_concentracion= sm.add_constant(azucar_concentracion["Concentración_gms"])
Y_azucar_concentracion= azucar_concentracion["θ_grados"]
modelos_azucar_concentracion = sm.OLS(Y_azucar_concentracion,X_azucar_concentracion).fit()
print(modelos_azucar_concentracion.summary())
```

```python
data = np.linspace(0, 100, 100)
plt.scatter(azucar_concentracion.Concentración_gms, azucar_concentracion.θ_grados, color = "black", label = "Datos")
plt.plot(data, 89.5656  +   0.1433 *data, c = "olive" ,label = "Ajuste", linestyle = "--")
plt.title("Actividad óptica azúcar: $\\lambda = 532nm$, 500ml $H_2 O$, L=8.944")
plt.xlabel("Concentración [gms]"); plt.ylabel("$\\theta[grados]$"); plt.grid(); plt.legend()
```

```python
azucar_camino_optico = pd.read_excel("Azucar_camino_optico_variable.xlsx")
X_azucar_camino_optico= sm.add_constant(azucar_camino_optico["camino_optico"])
Y_azucar_camino_optico= azucar_camino_optico["θ_grados"]
modelos_azucar_camino_optico = sm.OLS(Y_azucar_camino_optico, X_azucar_camino_optico).fit()
print(modelos_azucar_camino_optico.summary())
```

```python
data = np.linspace(7.8, 9, 100)
plt.scatter(azucar_camino_optico.camino_optico, azucar_camino_optico.θ_grados, color = "black", label = "Datos")
plt.plot(data,   122.6970    -3.2559*data, c = "olive" ,label = "Ajuste", linestyle = "--")
plt.title("Actividad óptica Azúcar camino optico: $\\lambda = 532nm$, 500ml $H_2 O$")
plt.xlabel("Camino óptico [cms]"); plt.ylabel("$\\theta[grados]$"); plt.grid(); plt.legend()
```
