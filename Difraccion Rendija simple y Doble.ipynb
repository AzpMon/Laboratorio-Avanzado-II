```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True

rendija_simple = pd.read_excel("rendija_simple.xlsx")
rendija_doble = pd.read_excel("rendija_doble.xlsx")
```

```python
columnas = ["Tiempo", "Voltaje"]
rendija_simple.columns = columnas
rendija_doble.columns = columnas


#Limpiamos los datos: Consideramos únicamente aquellos significativos para en análisis 
rendija_doble = rendija_doble[(rendija_doble.Tiempo > 3.5)]
rendija_doble = rendija_doble[(rendija_doble.Tiempo <5.6)]
rendija_simple = rendija_simple[(rendija_simple.Tiempo < 15 )]
rendija_simple = rendija_simple[(rendija_simple.Tiempo >6 )]
```

```python
#Se grafican la difracción de la doble rendija
plt.plot(rendija_doble.Tiempo, rendija_doble.Voltaje)
plt.title("Rendija doble")
plt.ylabel("Voltaje [Volts]"); plt.xlabel("Tiempo [segundos]");plt.grid()
```

```python
#Se Dividien los datos para obtener los respectivos máximos
t_3 = 3.7; t_2 = 4.1; t_1 = 4.5
max3 = rendija_doble.loc[rendija_doble['Tiempo'] <= t_3]
max2= rendija_doble.loc[(rendija_doble['Tiempo'] <= t_2) & (rendija_doble['Tiempo'] >t_3) ]
max1= rendija_doble.loc[(rendija_doble['Tiempo'] <= t_1) & (rendija_doble['Tiempo'] >t_2) ]

observaciones_max_3 = max3.loc[max3["Voltaje"]==max(max3["Voltaje"])]
observaciones_max_2 = max2.loc[max2["Voltaje"]==max(max2["Voltaje"])]
observaciones_max_1 = max1.loc[max1["Voltaje"]==max(max1["Voltaje"])]
observaciones_max = rendija_doble.loc[rendija_doble["Voltaje"]== max(rendija_doble["Voltaje"])]
```

```python
print(f'Para orden 3, el máximo corresponde a un tiempo {observaciones_max.Tiempo.mean()-observaciones_max_3.Tiempo.mean()}\n')
print(f'Para orden 2, el máximo corresponde a un tiempo {observaciones_max.Tiempo.mean()-observaciones_max_2.Tiempo.mean()}\n')
print(f'Para orden 1, el máximo corresponde a un tiempo {observaciones_max.Tiempo.mean()-observaciones_max_1.Tiempo.mean()}\n')
```
